```python
from skimage.metrics import structural_similarity as ssim
from sklearn.metrics import jaccard_score, f1_score
from scipy.ndimage import find_objects


def slice_signal(gt_img, pred_img, cell_mask, mode="gt", reducer="sum"):
    """
    mode: 'gt' (default), 'pred', or 'both' (gt+pred)
    reducer: 'sum' (default) or 'mean'
    """
    if reducer not in {"sum", "mean"}:
        raise ValueError("reducer must be 'sum' or 'mean'")
    if mode not in {"gt", "pred", "both"}:
        raise ValueError("mode must be 'gt', 'pred', or 'both'")

    gt_part  = gt_img * cell_mask
    pred_part = pred_img * cell_mask if pred_img is not None else None

    if mode == "gt":
        vol = gt_part
    elif mode == "pred":
        if pred_part is None:
            raise ValueError("pred_img required for mode='pred'")
        vol = pred_part
    else:  # both
        if pred_part is None:
            raise ValueError("pred_img required for mode='both'")
        vol = gt_part + pred_part

    if reducer == "sum":
        sig = vol.reshape(vol.shape[0], -1).sum(axis=1)
    else:
        # mean over only masked voxels per slice
        masked_counts = cell_mask.reshape(cell_mask.shape[0], -1).sum(axis=1).clip(min=1)
        sig = vol.reshape(vol.shape[0], -1).sum(axis=1) / masked_counts

    return sig  # shape (Z,)
    

def select_peak_window(gt_img, pred_img, cell_mask, k=16, mode="gt", reducer="sum"):
    sig = slice_signal(gt_img, pred_img, cell_mask, mode=mode, reducer=reducer)
    Z = sig.shape[0]
    if k >= Z:
        return np.arange(Z)

    # sliding window sum over signal
    cumsum = np.cumsum(np.r_[0, sig])
    win_sums = cumsum[k:] - cumsum[:-k]  # length Z-k+1
    start = int(np.argmax(win_sums))
    return np.arange(start, start + k)


def compute_metrics_on_indices(gt_img, seg, cell_mask, indices, binarize=True, data_range=1):
    ssim_vals, iou_vals, dice_vals = [], [], []

    if binarize:
        gt_mask  = (gt_img > 0).astype(np.uint8)
        seg_mask = (seg    > 0).astype(np.uint8)
    else:
        gt_mask  = gt_img.astype(float)
        seg_mask = seg.astype(float)

    gt_mask = gt_mask * cell_mask
    seg_mask = seg_mask * cell_mask

    for z in indices:
        gt_slice  = gt_mask[z]
        pred_slice = seg_mask[z]

        if np.any(gt_slice) or np.any(pred_slice):
            # SSIM needs 2D arrays
            ssim_val = ssim(gt_slice, pred_slice, data_range=data_range)

            # IoU/Dice on binary
            gt_bin = (gt_slice > 0).astype(np.uint8).ravel()
            pr_bin = (pred_slice > 0).astype(np.uint8).ravel()
            iou = jaccard_score(gt_bin, pr_bin, zero_division=0)
            dice = f1_score(gt_bin, pr_bin, zero_division=0)

            ssim_vals.append(ssim_val); iou_vals.append(iou); dice_vals.append(dice)

    return float(np.mean(ssim_vals)), float(np.mean(iou_vals)), float(np.mean(dice_vals))


def compute_metrics(gt_img, seg, cell_mask):
    best16_contig = select_peak_window(gt_img, seg, cell_mask, k=16, mode="gt", reducer="sum")
    ssim_m, iou_m, dice_m = compute_metrics_on_indices(gt_img, seg, cell_mask, best16_contig,
                                                   binarize=True, data_range=1)
    return ssim_m, iou_m, dice_m
```

```python
prediction_paths = [f"{BASE_PATH}/Inference/{organelle}/{path}" for path in os.listdir(f"{BASE_PATH}/Inference/{organelle}/")] # Channel 0
prediction_paths.sort(key=lambda x: int(x.split('.')[0].split('_')[-3]))
if organelle == "DNA" or organelle == "Plasma-membrane": # May be ok for all organelles
    base = "/sise/assafzar-group/assafzar/full_cells_fovs"
    gt_mask_paths = []
    for path in prediction_paths:
        org = path.split('/')[-1].split('_')[0]
        if org == "DNA":
            org = "Nucleolus-(Granular-Component)"
        num = path.split('/')[-1].split('_')[2]
        gt_mask_paths.append(f"{base}/{org}/{org}_{num}.tiff")
else:
    test_csv = pd.read_csv(test_csv_path)
    gt_mask_paths = [path for path in test_csv['path_tiff']] # Channel 6
    gt_mask_paths.sort(key=lambda x: int(x.split('.')[0].split('_')[-1]))
```

```python
# --- Process one sample ---
results = []
for i, (gt_path, pred_path) in enumerate(zip(gt_mask_paths, prediction_paths)):
    if i == 0:
        # Load images
        gt_image = tiff.imread(gt_path)
        cell_seg_mask = gt_image[5]
        if organelle == "DNA" or organelle == "Plasma-membrane":
            gt_seg_path = pred_path.replace('Inference', 'Segmentations')
            gt_seg_path = gt_seg_path.replace(f'{organelle}', f'{organelle}_GT', 1)
            gt_seg_path = gt_seg_path.replace('inference_outputs', 'gt_segmentation')
            gt_structure = tiff.imread(gt_seg_path)
            gt_structure = (gt_structure > 0).astype(np.uint8)
        else:
            gt_structure = gt_image[6]  # binary ground truth structure mask
        inference_image = tiff.imread(pred_path)
        confidence_map = inference_image[2]
        seg_path = pred_path.replace('Inference', 'Segmentations')
        seg_path = seg_path.replace('inference_outputs', 'predicted_segmentation')
        seg = tiff.imread(seg_path)
        seg_binary = (seg > 0).astype(np.uint8)
    
        # Move to the single cell level
        labels = np.unique(cell_seg_mask)
        labels = labels[labels != 0]
    
        for j, label in enumerate(labels):
            if j == 0:            
                # Create precise binary mask for current label
                mask = (cell_seg_mask == label)

                # Get bounding box of current cell mask
                coords = np.argwhere(mask)
                zmin, ymin, xmin = coords.min(axis=0)
                zmax, ymax, xmax = coords.max(axis=0) + 1  # +1 for slicing

                # Crop all inputs
                gt_crop = gt_structure[zmin:zmax, ymin:ymax, xmin:xmax]
                seg_crop = seg_binary[zmin:zmax, ymin:ymax, xmin:xmax]
                mask_crop = mask[zmin:zmax, ymin:ymax, xmin:xmax]
                conf_crop = confidence_map[zmin:zmax, ymin:ymax, xmin:xmax]
            
                # Check confidence
                masked_confidence = np.where(mask_crop, conf_crop, 0)
                max_error = max(masked_confidence[z].mean() for z in range(masked_confidence.shape[0]))

                # Calculate metrics
                # TODO: Fix segmentation function like AICS
                try:
                    ssim_score, iou, dice = compute_metrics(gt_crop, seg_crop, mask_crop)
                except Exception as e:
                    print(e)
                    ssim_score, iou, dice = np.nan, np.nan, np.nan
            
                results.append({
                    "index": i,
                    "cell": j,
                    "confidence": max_error,
                    "ssim": ssim_score,
                    "iou": iou,
                    "dice": dice
                })

# --- Create results DataFrame ---
df = pd.DataFrame(results)
df
```
